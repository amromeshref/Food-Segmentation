```python
import json
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the annotation data
with open("../data/external/public_validation_set_2.0/annotations.json", "r") as file:
    annotations = json.load(file)
```

```python
# Load the image (replace with actual path)
image_path = "../data/external/public_validation_set_2.0/images/123727.jpg"
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert to RGB for Matplotlib

ann_id = 174359

for ann in annotations["annotations"]:
    if ann["id"] == ann_id:
        segmentation = ann["segmentation"]

# Create a mask for visualization
mask = np.zeros_like(image)

# Draw the segmentations    
for segment in segmentation:
    points = np.array(segment, dtype=np.int32).reshape(-1, 2)  # Reshape to Nx2
    cv2.polylines(image, [points], isClosed=True, color=(255, 0, 0), thickness=2)
    cv2.fillPoly(mask, [points], color=(0, 255, 0))  # Fill mask with green

# Overlay the mask on the image
overlay = cv2.addWeighted(image, 0.7, mask, 0.3, 0)

# Show the image with segmentation
plt.figure(figsize=(10, 10))
plt.imshow(overlay)
plt.axis("off")
plt.show()
```
